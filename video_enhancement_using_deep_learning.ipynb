```python
#@title <h1>video upload</h1>
import os
from IPython.display import HTML, display, Image
from base64 import b64encode
from google.colab import files
from datetime import datetime

# Specify the custom path where you want to upload the file
source="video"
custom_path = "/content/"
out_file_path = ""
def rename_file_with_timestamp(original_file_name):
    # Get the current date and time in a specific format
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")

    # Extract file extension
    base_name, file_extension = os.path.splitext(original_file_name)

    # Construct the new file name
    new_file_name = f"{base_name}_BasetaTube_{timestamp}{file_extension}"

    return new_file_name
if source=="image":
  print("Source is image:")
  # Specify the custom path where you want to upload the file
  custom_path = "/content/"
  uploaded_files = files.upload()

  # Check if any files were uploaded
  if len(uploaded_files) == 0:
      print("No files uploaded. Please upload an image.")
  else:
      # Get the first uploaded file
      filename = next(iter(uploaded_files))

      # Check if the file has a valid image extension
      allowed_extensions = ['.png', '.jpg', '.jpeg']
      if not any(filename.lower().endswith(ext) for ext in allowed_extensions):
          print("Invalid file type. Please upload an image.")
      else:
          # Move the uploaded file to the custom path
          import os
          os.makedirs(custom_path, exist_ok=True)
          file_path = os.path.join(custom_path, filename)
          with open(file_path, 'wb') as f:
              f.write(uploaded_files[filename])

          # Encode the image file to base64
          image_base64 = b64encode(uploaded_files[filename]).decode('utf-8')

          # Display the uploaded image
          img_html = f'<img src="data:image/{filename.split(".")[-1]};base64,{image_base64}" alt="Uploaded Image" width="400">'
          display(HTML(img_html))

          # Save the file path in a variable for later use
          file_path
          out_file_path = out_file_path = file_path.replace(".", "_Baseta_Tube.")
          print("File Path:", file_path)
elif source =="video":
  print("Source is video:")
  # Create an HTML label with the text "Target video file"
  label_html = '<b><label for="target_video">Target video file:</label></b><br/><br/>'
  display(HTML(label_html))
  uploaded_files = files.upload()

  # Check if any files were uploaded
  if len(uploaded_files) == 0:
      print("No files uploaded. Please upload a video file.")
  else:
      # Get the first uploaded file
      filename = next(iter(uploaded_files))
      newfilename = rename_file_with_timestamp(filename)

      # Check if the file has a valid video extension
      allowed_extensions = ['.mp4', '.avi', '.mkv']
      if not any(filename.lower().endswith(ext) for ext in allowed_extensions):
          print("Invalid file type. Please upload a video file.")
      else:
          # Move the uploaded file to the custom path
          import os
          os.makedirs(custom_path, exist_ok=True)
          file_path = os.path.join(custom_path, filename)
          with open(file_path, 'wb') as f:
              f.write(uploaded_files[filename])
              new_file_name = rename_file_with_timestamp(filename)
              new_file_path = os.path.join(custom_path, new_file_name)
              os.rename(file_path, new_file_path)


          # Encode the video file to base64
          video_base64 = b64encode(uploaded_files[filename]).decode('utf-8')

          # Set the desired width and height
          width = 640
          height = 360

          # Create an HTML label with the text "Target video file"
          label_html = '<label for="target_video">Target video file:</label>'

          # Create an HTML tag with the embedded video and size attributes
          video_html = f'<video id="target_video" width="{width}" height="{height}" controls src="data:video/mp4;base64,{video_base64}" type="video/mp4"></video>'

          # Display the label and video
          display(HTML(label_html))
          display(HTML(video_html))

          # Save the file path in a variable for later use
          file_path = new_file_path
          out_file_path = out_file_path = file_path.replace(".", "_BasetaTube.")
          print("File Path:", new_file_path)
```
